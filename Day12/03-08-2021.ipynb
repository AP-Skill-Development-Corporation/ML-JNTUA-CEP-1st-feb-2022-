### Support vector machine
- Support vector machine is a linear model for both classification and regression. It can sovle linear and non-linear model problems
- It is based on the idea of finding a hyperplane or line which seperates the data into classes

#### Find Hyperplane
- The distance between hyperplane and nearest data point from either class is known as margin
- **Formula**:
    - w * x + b = -1
    - w * x + b = 0
    - w * x + b = +1

#### Predicting the lung cancer disease based on smoking and alcohol consumption

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/lung_cancer_examples.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
X = df.iloc[:,[2,3,4,5]]
Y = df['Result']
```

```python
X.head()
```

```python
from sklearn.svm import SVC
```

```python
svm1 = SVC()
svm1
```

```python
svm1 = SVC(kernel='linear')
svm1
```

```python
svm1.fit(X,Y)
```

```python
Y_predict1 = svm1.predict(X)
Y_predict1
```

```python
svm1.predict([[25,32,5,2]]) # True
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y,Y_predict1) # actual and predicted
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Student_PassOrFail.csv')
```

```python
df2.head()
```

```python
df2.isnull().sum().sum()
```

```python
df2.shape
```

```python
X2 = df2[['Self_Study_Daily','Tution_Monthly']] # features of student
Y2 = df2['Pass_Or_Fail'] # target
```

```python
svm2 = SVC(kernel='linear')
svm2
```

```python
svm2.fit(X2,Y2)
Y_predict2 = svm2.predict(X2)
Y_predict2
```

```python
accuracy_score(Y2,Y_predict2)
```

```python
svm3 = SVC(kernel='poly')
svm3
```

```python
svm3.fit(X2,Y2)
Y_predict3 = svm3.predict(X2)
```

```python
accuracy_score(Y2,Y_predict3)
```
