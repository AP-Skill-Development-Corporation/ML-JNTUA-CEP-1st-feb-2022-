```python
## Linear regression
X = boston_df[['RM']] ## Independent variable
Y = boston_df['cost'] ## Dependent or target
```

### Train / Test splitting of data
- Training set: This data is used to fit the model
- Testing set: Prediction or Evaluation of model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 45)
## test_size = 0.2 (20% of data for testing)
## random_state = non-zero. ( it will seperate data randomly)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
### Apply linear regression model
```

```python
from sklearn.linear_model import LinearRegression
## Y = mX + c
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
print('Training Score : ', model.score(X_train,Y_train))
print('Testing Score : ', model.score(X_test,Y_test))
```

- Above score is very low so need to improve the model.
